```python
saldo = float(input("Ingrese su saldo: "))
deposito = float(input("Ingrese el deposito que realizara: "))
retiro = float(input("Ingrese el retiro que realizara: "))

saldo += deposito
saldo -= retiro
saldo = saldo * (1 + 0.02)

print(f"Saldo final con interés: ${saldo:.2f}")
```

```python
# Lista inicial de temperaturas en °C
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]

# Agregar un temperatura nueva
temperaturas.append(26.3)
print("Agregar una temperatura", temperaturas)
# Eliminar la primera temperatuera
temperaturas.pop(0)
print("Eliminar una temperatura", temperaturas)
# Inserta una temperatura
temperaturas.insert(1, 23.7)
print("Insertar una temperatura en una posicion dada", temperaturas)
# Muestra la lista ordenada de menor a mayor
temperaturas.sort()
print("Mostrar lista ordenada", temperaturas)
# Calcular la diferencia
mayor = max(temperaturas)
menor = min(temperaturas)
diferencia = mayor - menor
promedio = sum(temperaturas) / len(temperaturas)
print(f"Diferencia: {diferencia:.2f}°C")
print(f"Promedio: {promedio:.2f}°C")
# 6. Encuentra el índice de la temperatura 23.7
indice_buscado = temperaturas.index(23.7)
print(f"El índice de 23.7 es: {indice_buscado}")

# 7. Muestra las temperaturas de la 2da a la 4ta posición
print("2da a la 4ta posición:", temperaturas[1:4])

# 8. Muestra las tres primeras temperaturas
print("Tres primeras:", temperaturas[:3])
```

```python
# Lista de productos en una tienda
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]
print(inventario[0])

nuevo = {"nombre": "Cereal", "precio": 2800, "stock": 12}
inventario.append(nuevo)
print(inventario)

for i in inventario:
    if i["nombre"] == "Leche":
        i["stock"] -= 2

print(inventario)

valor_total = sum(i["precio"] * i["stock"] for i in inventario)
print("Valor total del inventario:", valor_total)

for i in inventario:
    if i["stock"] < 10:
        print((f"- {i['nombre']}"))

for i in inventario:
    i["precio"] = i["precio"] * 1.10
print(inventario)

for i in inventario:
    i["precio"] = round(i["precio"], 2)

buscar = input("Ingrese el nombre del producto que desea buscar: ")
encontrado = True
for i in inventario:
    if i["nombre"] == buscar:
        print(f"Producto encontrado: {i}")
        encontrado = True
        break
if not encontrado:
    print("El producto no se encuentra en el inventario.")

inventario = [i for i in inventario if i["stock"] > 0]
print(inventario)

inventario.sort(key=lambda p: p["precio"])
print("Inventario final ordenado:", inventario)
```

```python
posiciones = [(0, 0), (1, 2), (3, 4)]

for i in range(len(posiciones)):

    x1 = posiciones[i-1][0]
    y1 = posiciones[i-1][1]

    x2 = posiciones[i][1]
    y2 = posiciones[i][1]

    distancia = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

print("La distancia recorrida es:", distancia)
```

```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

prestados = 0
disponibles = 0

for i in biblioteca["libros"]:
    if i["prestado"] == True:
        prestados += 1
    else:
        i["prestado"] == False
        disponibles += 1

print("Libros disponibles:", disponibles)
print("Libros prestados:", prestados)

nuevo = {"titulo": " Don Quijote de la Mancha", "autor": "Miguel de Cervantes", "prestado": False}
biblioteca["libros"].append(nuevo)
print("Agregar un libro")
print(biblioteca)

for i in biblioteca["libros"]:
    if i["titulo"] == "1984":
        i["prestado"] = True

print("Modificacion de estado")
print(biblioteca)
```
