```python
saldo = float(input("Ingrese su saldo: "))
deposito = float(input("Ingrese el deposito que realizara: "))
retiro = float(input("Ingrese el retiro que realizara: "))

saldo += deposito
saldo -= retiro
saldo = saldo * (1 + 0.02)

print(f"Saldo final con interés: ${saldo:.2f}")
```

```python
# Lista inicial de temperaturas en °C
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]

# Agregar un temperatura nueva
temperaturas.append(26.3)
print("Agregar una temperatura", temperaturas)
# Eliminar la primera temperatuera
temperaturas.pop(0)
print("Eliminar una temperatura", temperaturas)
# Inserta una temperatura
temperaturas.insert(1, 23.7)
print("Insertar una temperatura en una posicion dada", temperaturas)
# Muestra la lista ordenada de menor a mayor
temperaturas.sort()
print("Mostrar lista ordenada", temperaturas)
# Calcular la diferencia
mayor = max(temperaturas)
menor = min(temperaturas)
diferencia = mayor - menor
promedio = sum(temperaturas) / len(temperaturas)
print(f"Diferencia: {diferencia:.2f}°C")
print(f"Promedio: {promedio:.2f}°C")
# 6. Encuentra el índice de la temperatura 23.7
indice_buscado = temperaturas.index(23.7)
print(f"El índice de 23.7 es: {indice_buscado}")

# 7. Muestra las temperaturas de la 2da a la 4ta posición
print("2da a la 4ta posición:", temperaturas[1:4])

# 8. Muestra las tres primeras temperaturas
print("Tres primeras:", temperaturas[:3])
```
